# Preprocessing

```python
# missing value를 없애는게 목적 No.1
total = train.isnull().sum().sort_values(ascending=False) #missing value에 대한 true false구분
percent = (train.isnull().sum()/train.isnull().count()).sort_values(ascending=False) #더함
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(25)
```

```python
train = train.drop((missing_data[missing_data['Total'] > 81]).index, axis=1)
# missing data가 82개 이상이면 버림, Lotfrontage 삭제
```

```python
train.isnull().sum().sort_values(ascending=False).head(20)
```

```python
train.head()
```

```python
# fillna : NAN을 fill한다 mean값으로
train['GarageYrBlt'] = train['GarageYrBlt'].fillna(train['GarageYrBlt'].mean())
train['MasVnrArea'] = train['MasVnrArea'].fillna(train['MasVnrArea'].mean())
```

# Build a sparse linear model

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    train.drop('SalePrice', axis=1), train['SalePrice'],
    test_size=0.3, random_state=101)
```

```python
y_train= y_train.values.reshape(-1,1)
y_test= y_test.values.reshape(-1,1)

from sklearn.preprocessing import StandardScaler
standard_X = StandardScaler()
standard_y = StandardScaler()
X_train = standard_X.fit_transform(X_train)
X_test = standard_X.transform(X_test)
y_train = standard_y.fit_transform(y_train)
y_test = standard_y.transform(y_test)
```

```python
from sklearn import linear_model
# Lasso model
model = linear_model.Lasso(alpha = 0.1)
model.fit(X_train, y_train)
```

# Prediction

```python
# Lasso Prediction
lasso_predict = model.predict(X_test)
print(lasso_predict.shape)
lasso_predict = lasso_predict.reshape(-1,1)
y_test = y_test.reshape(-1,1)
print(y_test.shape)
```

```python
# draw plot for predictions and test values
plt.figure(figsize = (15,8))
plt.scatter(y_test,lasso_predict)
plt.xlabel('Real')
plt.ylabel('Lasso Prediction')
```

```python
from sklearn import metrics

# Eval
print('MAE : ', metrics.mean_absolute_error(y_test, lasso_predict))
print('MSE : ', metrics.mean_squared_error(y_test, lasso_predict))
print('RMSE : ', np.sqrt(metrics.mean_squared_error(y_test, lasso_predict)))
```
